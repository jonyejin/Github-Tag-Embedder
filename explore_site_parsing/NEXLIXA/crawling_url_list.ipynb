## Load data

```python
import requests
from bs4 import BeautifulSoup as bs
import json

urls = ["https://github.com/topics/python"]

prettyContent = ""
dict = {}

for url in urls:

    arr = []
    response = requests.get(url)
    key = url.replace("https://github.com/topics/","")
    with open("raw_python.html") as fp:
        soup = bs(fp, 'html.parser')
    prettyContent = soup.prettify()
    
    with open('./0_raw.txt', 'w') as file:
        file.write(prettyContent)
# https://www.crummy.com/software/BeautifulSoup/bs4/doc/

    with open('./1_repository_card.txt', 'w') as dst:
        with open("/Users/andrewwonwhoonah/Github-Tag-Embedder/explore_site_parsing/NEXLIXA/0_raw.txt", 'r') as f:
            while True:
                line = f.readline()
                if not line:
                    break
                if "REPOSITORY_CARD" in line:
                    dst.write(line)

    with open('/Users/andrewwonwhoonah/Github-Tag-Embedder/explore_site_parsing/NEXLIXA/2_originating_url.txt', 'w') as dst:
        with open("/Users/andrewwonwhoonah/Github-Tag-Embedder/explore_site_parsing/NEXLIXA/1_repository_card.txt", 'r') as f:
            while True:
                line = f.readline()
                if not line:
                    break
                else:
                    # string에 :을 넣으면 안 됨
                    dst.write(line[line.find("data-hydro-click-hmac"):])
                
    # href 뒤로만 찾기
    # topic 들어간건 다 버리고
    with open('./3_repository_href.txt', 'w') as dst:
        with open("/Users/andrewwonwhoonah/Github-Tag-Embedder/explore_site_parsing/NEXLIXA/2_originating_url.txt", 'r') as f:
            while True:
                line = f.readline()
                if not line:
                    break
                if "topics" in line:
                    continue
                if "href=" in line:
                    dst.write(line[line.find("href=") + 6:-3])
                    dst.write("\n")
    tmp_name = "/keras-team"

    with open('./4_repository.txt', 'w') as dst:
        with open("/Users/andrewwonwhoonah/Github-Tag-Embedder/explore_site_parsing/NEXLIXA/3_repository_href.txt", 'r') as f:
            while True:
                line = f.readline()
                if not line:
                    break
                if "id" in line:
                    continue
                elif "style" in line:
                    continue
                else:
                    dst.write(line)            

    with open("/Users/andrewwonwhoonah/Github-Tag-Embedder/explore_site_parsing/NEXLIXA/4_repository.txt", 'r') as f:
        while True:
            line = f.readline()
            if not line:
                break
            if line.count("/") == 4:
                # line = "https://github.com" + line
                arr.append(line[:-1])
    dict[key] = arr

with open("/Users/andrewwonwhoonah/Github-Tag-Embedder/explore_site_parsing/NEXLIXA/output.json", "w") as json_file:
    json.dump(dict, json_file)



```
